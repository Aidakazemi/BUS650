## üß© Part 1 ‚Äî Population vs Sample

**Question:**  
> ‚ÄúIf we wanted to know the average age of all MBA students in Canada, do we have to ask everyone?‚Äù

**Concepts**
- **Population** ‚Üí the entire group we care about (e.g., *all MBA students in Canada*).  
- **Sample** ‚Üí a smaller group we actually collect data from (e.g., *our BUSI650 class*).  

We use the **sample** to make conclusions about the **population** ‚Äî the foundation of business analytics.

**Business Analogy:**  
A marketing team surveys **200 customers** (sample) to predict behavior of **all customers** (population).

## üßÆ Step 3 ‚Äî One-Sample t-Test

Now we test a simple question:

> ‚ÄúIs the average age of our BUSI650 class different from 30 years old?‚Äù

This is a **one-sample t-test** ‚Äî we test whether our sample mean differs from a hypothesized population mean.

---

### Hypotheses

| Symbol | Meaning | Statement |
|:--:|:--|:--|
| $H_0$ | Null hypothesis | $\mu = 30$ (no difference) |
| $H_1$ | Research hypothesis | $\mu \neq 30$ (mean age is different from 30) |

---

### Data

- Sample mean ($\bar{x}$) = 29.3  
- Sample SD ($s$) = 2.38  
- Sample size ($n$) = 10  
- Hypothesized mean ($\mu_0$) = 30  

---

### Step 3.1 ‚Äî Compute the t-statistic

Formula:

$$
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
$$

$$
t = \frac{29.3 - 30}{2.38 / \sqrt{10}}
= \frac{-0.7}{2.38 / 3.16}
= \frac{-0.7}{0.75}
= -0.93
$$

---

### Step 3.2 ‚Äî Compare to Critical Value

Degrees of freedom: $df = n - 1 = 9$  
At $\alpha = 0.05$, two-tailed, $t_{critical} ‚âà ¬±2.26$

| t-value | t-critical | Decision |
|:--:|:--:|:--:|
| -0.93 | ¬±2.26 | Fail to reject $H_0$ |

**Conclusion:**  
> The average age of our sample (29.3) is *not significantly different* from 30.  
> Any small difference is likely due to random variation.

## üß≠ Step 4 ‚Äî One-Tailed and Two-Tailed Tests (Group Comparison)

Let‚Äôs revisit the example we used last week ‚Äî comparing the **average age of male and female students** in our class.

---

### Data (small sample for demonstration)

| Gender | Ages |
|:--:|:--:|
| Male | 32, 34, 31, 35, 29 |
| Female | 28, 30, 26, 27, 31 |

---

### Step 4.1 ‚Äî Compute Group Statistics

| Group | n | Mean | SD |
|:--:|:--:|:--:|:--:|
| Male | 5 | 32.2 | 2.39 |
| Female | 5 | 28.4 | 2.07 |

---

### Step 4.2 ‚Äî Hypotheses

| Symbol | Meaning | Statement |
|:--:|:--|:--|
| $H_0$ | Null Hypothesis | $\mu_{male} = \mu_{female}$ |
| $H_1$ | Research Hypothesis | $\mu_{male} > \mu_{female}$ (men are older) |

This is a **one-tailed test** since we expect a direction.

---

### Step 4.3 ‚Äî Calculate the t-statistic

Pooled SD:

$$
s_p = \sqrt{\frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1 + n_2 - 2}}
$$

$$
s_p = \sqrt{\frac{(4)(2.39)^2 + (4)(2.07)^2}{8}} = \sqrt{\frac{22.84 + 17.13}{8}} = \sqrt{4.99} = 2.23
$$

t-statistic:

$$
t = \frac{\bar{x}_1 - \bar{x}_2}{s_p\sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}
$$

$$
t = \frac{32.2 - 28.4}{2.23 \times \sqrt{0.4}} = \frac{3.8}{2.23 \times 0.632} = \frac{3.8}{1.41} = 2.69
$$

df = 8  
$t_{critical}(0.05)$ (one-tailed) ‚âà 1.86  

Since **2.69 > 1.86**, we **reject $H_0$**.

---

### Step 4.4 ‚Äî Interpretation

> There is enough evidence to suggest that male students in this class are significantly older than female students.

**In plain words:**  
> ‚ÄúIf we took many random samples like this, we would rarely see a gap this large if men and women had the same average age.‚Äù

---

## üí¨ Step 5 ‚Äî One-Tailed vs Two-Tailed (Philosophy Recap)

| Test Type | When Used | Example | $H_1$ |
|:--|:--|:--|:--|
| **One-tailed** | Expect change in one direction | ‚ÄúMen are older than women.‚Äù | $\mu_{male} > \mu_{female}$ |
| **Two-tailed** | Any difference possible | ‚ÄúMen and women differ in age.‚Äù | $\mu_{male} \neq \mu_{female}$ |

---

## ‚úÖ Summary

| Concept | Meaning |
|----------|----------|
| Population vs Sample | The group we want to understand vs. the group we actually measure |
| Descriptive Stats | Summarize data: mean, median, mode, variance, SD |
| One-sample t-test | Compares one sample mean to a known or claimed value |
| Two-sample t-test | Compares means between two independent groups |
| One-tailed | Predicts direction (e.g., greater than) |
| Two-tailed | Detects any difference (either higher or lower) |

## üßÆ Step 2 ‚Äî The Big Idea

In simple language:

> Regression draws a **line** through your data that best summarizes the overall pattern.

That line is described by the equation:

$$
\hat{Y} = a + bX
$$

where:

| Symbol | Meaning | Explanation |
|:--:|:--|:--|
| $\hat{Y}$ | Predicted value | The outcome we estimate (e.g., predicted sales) |
| $a$ | Intercept | Value of Y when X = 0 (baseline) |
| $b$ | Slope | Change in Y for each 1-unit change in X |
| $X$ | Predictor (independent variable) | What we think influences Y |
| $Y$ | Actual dependent variable | The variable we want to explain or predict |

**Intuition:**  
- $b$ shows *direction* and *strength* of relationship.  
- If $b > 0$, Y increases as X increases (positive relationship).  
- If $b < 0$, Y decreases as X increases (negative relationship).

---

## üß† Step 3 ‚Äî Business Example

Let‚Äôs take a simple dataset about **advertising and sales**.

A small retail company collected data on how much they spent on advertising (in $1,000s) and how much sales (in $
10,000s) they made that month.

| Advertising ($X$) | Sales ($Y$) |
|--------------------|-------------|
| 1 | 4 |
| 2 | 6 |
| 3 | 7 |
| 4 | 10 |
| 5 | 12 |

---

### ‚úèÔ∏è What We Want to Know

> ‚ÄúIf we increase our advertising budget, how much will sales go up?‚Äù

To answer that, we fit a line that **best represents** how Sales (Y) change with Advertising (X).

---

## üßÆ Step 4 ‚Äî The Regression Equation

We fit this line:

$$
\hat{Y} = a + bX
$$

where $a$ and $b$ are chosen to minimize the total error between the actual data and predicted values.

**Error (residual)** = $Y - \hat{Y}$

The line that makes those errors as small as possible is called the **line of best fit** (Least Squares Line).

---

### üßæ Let‚Äôs Calculate It in Python


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simple data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([4, 6, 7, 10, 12])

# Create and fit regression model
model = LinearRegression()
model.fit(X, Y)

a = model.intercept_
b = model.coef_[0]

print(f"Regression Equation: ≈∂ = {a:.2f} + {b:.2f}X")

# Plot the data and regression line
plt.scatter(X, Y, color="blue", label="Actual Data")
plt.plot(X, model.predict(X), color="red", label="Regression Line")
plt.xlabel("Advertising ($000s)")
plt.ylabel("Sales ($10,000s)")
plt.title("Sales vs Advertising Spend")
plt.legend()
plt.show()
```

Regression Equation: ≈∂ = 2.20 + 1.90X


### üß© Step 5 ‚Äî Interpretation

The equation is:

$$
\hat{Y} = 2.20 + 1.90X
$$

**Interpretation:**
- **Intercept (2.20):** When advertising is 0, sales are around $2.2 √ó 10,000 = $22,000 (baseline).  
- **Slope (1.90):** For every extra $1,000 spent on advertising, sales increase by about $19,000.  

**In plain language:**  
> ‚ÄúAdvertising seems to have a strong, positive impact on sales.‚Äù

- The **slope** shows how fast Y grows as X increases.
